```python
import math

def f1(x):
    return x*x*x/3 - 2*x*x + 2*x +3

def f2(x):
    return x*x + math.sin(x)

def f3(x):
    return x*x*x*x - 6*x*x*x + 13*x*x - 12*x + 4

def f4(x):
    return 2*x*x*x - 3 * x + 15
```

```python
def deriv_g1(f, x, h):
    elem1 = 3 * f(x) - 4 * f(x - h) + f(x - 2*h)
    elem2 = 2 * h
#     print("DERIV")
#     print(f(x))
#     print(f(x-h))
#     print(f(x-2*h))
#     print(elem1)
#     print()
    return elem1/elem2

def deriv_g2(f, x, h):
    elem1 = -f(x+2*h) + 8*f(x+h)-8*f(x-h)+f(x-2*h)
    elem2 = 12*h
    return elem1/elem2
```

```python
def dehghan_hajarian(f, x0, metoda_derivare, epsilon, kmax, h):
    xk = x0
    k = 0
    
    deriv_xk = metoda_derivare(f, xk, h)
    deriv_gxk = metoda_derivare(f, xk + deriv_xk, h)
    
    if abs(deriv_gxk - deriv_xk) <= epsilon:
        return xk, k
    
#     print("g(x)" + str(deriv_xk))
#     print("g(g(x)+x)" + str(deriv_gxk))
#     print(abs(deriv_gxk - deriv_xk))
    z = xk + (deriv_xk * deriv_xk) / (deriv_gxk - deriv_xk)
    print("z", str(z))
    delta_xk = (deriv_xk * (metoda_derivare(f, z, h) - deriv_xk)) / (deriv_gxk- deriv_xk)
    
    x_next = xk - delta_xk
    xk = x_next
    k += 1

#     print("delta" , str(delta_xk))
    while abs(delta_xk) >= epsilon and k<=kmax and abs(delta_xk) <= 10**8:
        deriv_xk = metoda_derivare(f, xk, h)
        deriv_gxk = metoda_derivare(f, xk + deriv_xk, h)

        if abs(deriv_gxk - deriv_xk) <= epsilon:
            return xk, k

        z = xk + (deriv_xk * deriv_xk) / (deriv_gxk - deriv_xk)
        delta_xk = deriv_xk * (metoda_derivare(f, z, h) - deriv_xk) / (deriv_gxk - deriv_xk)
#         print(delta_xk)

        x_next = xk - delta_xk
        xk = x_next
        k += 1
        
    
        
    if abs(delta_xk) < epsilon:
        if abs(deriv_gxk- deriv_xk) > 10: #inseamna ca delta e 0 din alte cauze
            return "divergent", kmax
        else:
            return xk, k
    else:
        return "divergent", k
    
    
```
