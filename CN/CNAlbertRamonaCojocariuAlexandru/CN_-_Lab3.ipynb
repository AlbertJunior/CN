```python
def create_empty_matrix(n):
    empty_matrix = []
    for i in range(n):
        empty_matrix += [dict()]
    return empty_matrix

def citeste_matrice_rara(fileName):
    n = 0
    empty_matrix = None
    with open(fileName) as f:
        line_number = 0
        for line in f.readlines():
            line = line[:-1]
            line_number += 1
            if line_number == 1:
                n =  int(line)
                empty_matrix = create_empty_matrix(n)
            elif line_number == 2:
                continue
            else:
                value, i, j = line.split(", ")
                value = float(value)
                i = int(i)
                j = int(j)
                if j in empty_matrix[i].keys():
                    empty_matrix[i][j] += value
                else:
                    empty_matrix[i][j] = value
                
    return empty_matrix


matrix = citeste_matrice_rara("a.txt")
# matrix
```

```python
import copy

def adunareDictionar(matrice, linie, cheie, valoare):
    if cheie in matrice[linie].keys():
        matrice[linie][cheie] += valoare
    else:
        matrice[linie][cheie] = valoare
    return matrice


def suma(matrix, matrice_tridiagonala):
    a, b, c, p, q = matrice_tridiagonala
    suma = copy.deepcopy(matrix)
    n = len(a)
    
    adunareDictionar(suma, 0, 0, a[0]);
    adunareDictionar(suma, 0, 1, b[0]);
    
    i = 0
    for aa, bb, cc in zip(a[1:-1], b[1:], c[:-1]):
        i += 1
        adunareDictionar(suma, i, i-1, cc);
        adunareDictionar(suma, i, i, aa);
        adunareDictionar(suma, i, i+1, bb);
        
    adunareDictionar(suma, n-1, n-1, a[-1]);    
    adunareDictionar(suma, n-1, n-2, c[-1]);        
    return suma


def sumaBonus(matrix, matrice_tridiagonala):
    a, b, c, p, q = matrice_tridiagonala
    suma = copy.deepcopy(matrix)
    n = len(a)
    jB = p
    jC = - q
    countBB = 0
    countBC = 0
    for i in range(n):
        if a[i] > 0:
            adunareDictionar(suma, i, i, a[i])
        if jB < n:
            if b[countBB] > 0:
                adunareDictionar(suma, i, jB, b[countBB])
            countBB += 1
        if jC >= 0:
            if  c[countBC] > 0:
                adunareDictionar(suma, i, jC, c[countBC])
            countBC += 1
        jB += 1
        jC += 1
        
    return suma

    
```

```python
def check(matrix1, matrix2, epsilon):
    sum = 0
    for i in range(len(matrix1)):
        keys = list(matrix1[i].keys()) + list(matrix2[i].keys())
        keys = set (keys)
        for key in keys:
            value1 = matrix1[i].get(key, 0)
            value2 = matrix2[i].get(key, 0)
#             print(value1, value2, "indici", i, key)
            sum += abs(value1-value2)
    print(sum)
    return sum < epsilon, sum
```

```python
def prod_boria(matrix, matrice_tridiagonala):
    a, b, c, p, q = matrice_tridiagonala
    prod = []
    n = len(a)
    
#     prima linie
    prod += [dict()]
    for j in range(n):
        value = 0
        value += matrix[0].get(j, 0) * a[0]
        value += matrix[1].get(j, 0) * b[0]
        if value > 0:
            prod[-1][j] = value
     
    i = 0
    for aa, bb, cc in zip(a[1:-1], b[1:], c[:-1]):
        i += 1
        prod += [dict()]
        for j in range(n):
            value = 0
            value += matrix[i-1].get(j, 0) * cc
            value += matrix[i].get(j, 0) * aa
            value += matrix[i+1].get(j, 0) * bb 
            if value > 0:
                prod[-1][j] = value
        
            
#     ultima linie
    prod += [dict()]
    for j in range(n):
        value = 0
        value += matrix[-2].get(j, 0) * c[-1]
        value += matrix[-1].get(j, 0) * a[-1]
        if value > 0:
            prod[-1][j] = value
            
    return prod
    
    
def prod_aorib(matrix, matrice_tridiagonala):
    a, b, c, p, q = matrice_tridiagonala
    prod = []
    n = len(a)
    for i in range(n):
        prod += [dict()]  
#         prima coloana    
        value = 0
        value += matrix[i].get(0, 0) * a[0]
        value += matrix[i].get(1, 0) * c[0]
        if value > 0:
            prod[-1][0] = value
    
        j = 0
        for aa, bb, cc in zip(a[1:-1], b[:-1], c[1:]):
            j += 1
            
            value = 0
            value += matrix[i].get(j-1, 0) * bb
            value += matrix[i].get(j, 0) * aa
            value += matrix[i].get(j+1, 0) * cc 
            if value > 0:
                prod[-1][j] = value
            
        value = 0
        value += matrix[i].get(n-2, 0) * b[-1]
        value += matrix[i].get(n-1, 0) * a[-1]
        if value > 0:
            prod[-1][n-1] = value
        
    return prod
        

def prod_aorib_bonus(matrix, matrice_tridiagonala):
    a, b, c, p, q = matrice_tridiagonala
    prod = []
    n = len(a)
    for i in range(n):
        prod += [dict()]
    
        jB = 1 - p
        jC = 1 + q
        countBB = 0
        countBC = 0
        for idy, elemB in enumerate(a):
            if matrix[i].get(idy, 0) * elemB > 0:
                adunareDictionar(prod, i, idy, matrix[i].get(idy, 0) * elemB)
            if jB > 0:
                if matrix[i].get(jB-1, 0) * b[countBB] > 0:
                    adunareDictionar(prod, i, idy, matrix[i].get(jB-1, 0) * b[countBB])
                countBB += 1
            if jC <= n:
                if matrix[i].get(jC-1, 0) * c[countBC] > 0:
                    adunareDictionar(prod, i, idy, matrix[i].get(jC-1, 0) * c[countBC])
                countBC += 1
                
            jB += 1
            jC += 1
            
    return prod
    
```

```python
def bonus_tema3(matrice_tridiagonalaA, matrice_tridiagonalaB):
    aA, bA, cA, pA, qA = matrice_tridiagonalaA
    aB, bB, cB, pB, qB = matrice_tridiagonalaB
    n = len(aA)
    empty_matrix = create_empty_matrix(n)
    iA = 1
    iB = 1 + pA
    iC = 1 - qA    
    countAB = 0
    countAC = 0
    for idx, elemA in enumerate(aA):
        jA = 1
        jB = 1 - pB
        jC = 1 + qB
        countBB = 0
        countBC = 0
        for idy, elemB in enumerate(aB):
            if iA == jA:
                empty_matrix =  adunareDictionar(empty_matrix, idx, idy, elemA*elemB)
            if jB > 0 and iA==jB:
                empty_matrix =  adunareDictionar(empty_matrix, idx, idy, elemA*bB[countBB])
            if jC <= n and iA==jC:
                empty_matrix =  adunareDictionar(empty_matrix, idx, idy, elemA*cB[countBC])
                         
            if iB <= n:
                if iB == jA:
                    empty_matrix =  adunareDictionar(empty_matrix, idx, idy, bA[countAB]*elemB)
                if jB > 0 and iB==jB:
                    empty_matrix =  adunareDictionar(empty_matrix, idx, idy, bA[countAB]*bB[countBB])
                if jC <= n and iB==jC:
                    empty_matrix =  adunareDictionar(empty_matrix, idx, idy, bA[countAB]*cB[countBC])
                    
            if iC > 0:
                if iC == jA:
                    empty_matrix =  adunareDictionar(empty_matrix, idx, idy, cA[countAC]*elemB)
                if jB > 0 and iC==jB:
                    empty_matrix =  adunareDictionar(empty_matrix, idx, idy, cA[countAC]*bB[countBB])
                if jC <= n and iC==jC:
                    empty_matrix =  adunareDictionar(empty_matrix, idx, idy, cA[countAC]*cB[countBC])
                    
            if jB > 0:
                countBB += 1
            if jC < n:
                countBC += 1
            jA += 1
            jB += 1
            jC += 1
            
        if iC > 0: 
            countAC += 1
        if iB < n:
            countAB += 1
        iA += 1
        iB += 1
        iC += 1

    return empty_matrix
```

```python
def transform(matrice_tridiagonala):
    # pentru a transforma dintr-o matrice tridiagonala in matrice rara
    # am inmultit matricea cu matricea I
    a, b, c, p, q = matrice_tridiagonala
    n = len(a)
    empty_matrix = create_empty_matrix(n)
    for i in range(n):
        empty_matrix[i][i]=1
    rare_matrix = prod_aorib_bonus(empty_matrix, matrice_tridiagonala)
    
    return rare_matrix
```
